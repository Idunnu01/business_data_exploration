```python
# Load the dataset
file_path = '/Users/idunnuomisore/Documents/Projects/business-data-exploration-python/data/superstore_sales.csv'
data = pd.read_csv(file_path)

# Display basic information
print("Dataset Head:")
data.head()
```

```python
print("\nDataset Info:")
data.info()
```

```python
print("\nMissing Values:")
print(data.isnull().sum())
```

```python
print("\nDescriptive Statistics:")
data.describe()
```

```python
# Calculate mean, median, variance for key numeric columns
print("\nMean of Sales:")
print(data['Sales'].mean())

print("\nMedian of Sales:")
print(data['Sales'].median())

print("\nVariance of Sales:")
print(data['Sales'].var())
```

```python
# Group sales by category
print("\nSales by Category:")
print(data.groupby('Category')['Sales'].sum())

# Group sales by ship mode
print("\nSales by Ship mode:")
print(data.groupby('Ship Mode')['Sales'].sum())

# Group sales by Segment
print("\nSales by Segment:")
print(data.groupby('Segment')['Sales'].sum())

# Group sales by Country
print("\nSales by Country")
print(data.groupby('Country')['Sales'].sum())
```

```python
# Convert 'Order Date' to datetime with day-first format
data['Order Date'] = pd.to_datetime(data['Order Date'], dayfirst=True)

# Aggregate sales over time (e.g., monthly sales)
monthly_sales = data.resample('M', on='Order Date')['Sales'].sum()

# Line chart for monthly sales
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-', color='blue')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)
plt.tight_layout()
plt.savefig("/Users/idunnuomisore/Documents/Projects/business-data-exploration-python/visuals/sales_line_chart.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert 'Order Date' to datetime with day-first format
data['Order Date'] = pd.to_datetime(data['Order Date'], dayfirst=True)

# Extract year and month for grouping
data['Year'] = data['Order Date'].dt.year
data['Month'] = data['Order Date'].dt.month

# Group data by Year and Month, summing the Sales
monthly_sales = data.groupby(['Year', 'Month'])['Sales'].sum().reset_index()

# Pivot the table to have 'Month' as the x-axis and 'Year' as columns
pivot_sales = monthly_sales.pivot(index='Month', columns='Year', values='Sales')

# Plot the multi-line chart
plt.figure(figsize=(10, 6))
for year in pivot_sales.columns:
    plt.plot(pivot_sales.index, pivot_sales[year], marker='o', linestyle='-', label=str(year))

# Customize the plot
plt.title("Monthly Sales Trend by Year")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(range(1, 13), [
    "Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
])
plt.legend(title="Year")
plt.grid(True)
plt.tight_layout()

# Save the chart
plt.savefig("/Users/idunnuomisore/Documents/Projects/business-data-exploration-python/visuals/sales_trend_by_year.png")
plt.show()
```
